```python
test_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap=0)
texts = test_splitter.split_documents(data)
```

```python
embeddings = OpenAIEmbeddings(openai_api_key="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")
```

```python
pinecone.init(
    api_key = "f455d8e0-0fc6-4520-9681-b15bb5d466fd",
    environment = "asia-southeast1-gcp-free"
)
index_name = "abc-client-corporation"
```

```python
doc_search = Pinecone.from_texts([t.page_content for t in texts], embedding=embeddings, index_name=index_name)
```
