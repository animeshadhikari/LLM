```python
#from openai.api_resources import engine
# Take query and generate embedding
limit = 3050
def construct_prompt(query, embedding_model):
  embed_query = openai.Embedding.create(
      input = query,
      engine = embedding_model
  )
  # Retrieve from pinecone
  query_embeds = embed_query['data'][0]['embedding']

  # Get relavant contexts
  response = index.query(query_embeds, top_k=3, include_metadata=True)
  contexts = [x['metadata']['text'] for x in response['matches']]

  # Build our prompt with retrieved contexts included
  prompt_start = ("Answer the question baseed on the context below.\n\n Context:\n")
  prompt_end = (f"\n\nQuestion: {query}\n Answer:")

  #Append contexts untill hitting limit
  for i in range(len(contexts)):
    if len("-".join(contexts[:i])) >= limit:
      prompt = (prompt_start + "-".join(contexts[:i-1])+prompt_end)
      break
    elif i == len(contexts)-1:
      prompt = (prompt_start + "-".join(contexts)+prompt_end)
  return prompt
```

```python
prompt_with_context = construct_prompt(query, embedding_model)
prompt_with_context
```

```python
response = openai.Completion.create(
    engine = 'text-davinci-003',
    prompt = prompt_with_context,
    temperature = 0,
    max_tokens = 350,
    top_p=1
)
response['choices'][0]['text']
```
